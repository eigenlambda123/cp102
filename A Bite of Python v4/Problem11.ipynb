```python
# Accept student name and grade and save it to a dictionary.
# Display the content of the dictionary.
# Display the name  and grade of student with the highest and lowest grade.

data = {}

while True:
    print("""
User Choices:
1. Add Student and Grade
2. Edit Student Grade
3. Show All Students and Grade
4. Calculate GPA
5. Exit
    """)
    user_input = int(input())

    if user_input == 5:
        print('Program Terminated')
        break

    # Calculate GPA
    elif user_input == 4:
        print("""
Calculate GPA:
1. All Students
2. Specific Student
            """)
        gpa_choice = int(input())
        
        if gpa_choice == 1:
            for student, grades in data.items():
                gpa = sum(grades) / len(grades)
                print(f"{student}'s GPA: {gpa:.2f}")
        
        elif gpa_choice == 2:
            student_name = input("Enter student name: ")
            if student_name in data:
                grades = data[student_name]
                gpa = sum(grades) / len(grades)
                print(f"{student_name}'s GPA: {gpa:.2f}")
            else:
                print(f"Student {student_name} not found.")
                
    # Display Grade
    elif user_input == 3:
        print("All Students and Grades:")
        for student, grades in data.items():
            print(f"{student}: {grades}")

    # Edit Student
    elif user_input == 2:
        edit_student = input("Enter student name you want to edit: ")
        if edit_student in data:
            print(f"Current grades for {edit_student}: {data[edit_student]}")
            new_grades = []
            subject_count = len(data[edit_student])
            for i in range(1, subject_count + 1):
                grade = eval(input(f"Enter new grade for subject #{i}: "))
                new_grades.append(grade)
            grade[edit_student] = new_grades
            print(f"Grades updated for {edit_student}: {data[edit_student]}")
        else:
            print(f"Student {edit_student} not found.")

    # Add Student
    elif user_input == 1:
        while True:
            name = input("Enter student name: ")
            subject = int(input("How many subjects does the student have: "))
            grade = []
            for i in range(1, subject + 1):
                x = eval(input("Enter " + name +"'s grade for subject #" + str(i) + ": "))
                grade.append(x)
            data[name] = grade
            
            cont = input("Do you want to add more students? [y/n]: ").lower()
            if cont == 'n':
                break
            else:
                continue
```
