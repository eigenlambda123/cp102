```python
# Create a Python program that will demonstrate list comprehension.

def add_vectors(v1, v2):
    if len(v1) != len(v2):
        raise ValueError("Vectors must be the same length.")
    return [v1[i] + v2[i] for i in range(len(v1))]

def scalar_matrix(scalar, vector):
    return [scalar * x for x in vector]

def multiply_matrix(m1, m2):
    if len(m1[0]) != len(m2):
        raise ValueError("Number of columns in the first matrix must equal the number of rows in the second matrix.")
    return [
        [
            sum(m1[i][k] * m2[k][j] for k in range(len(m2)))
            for j in range(len(m2[0]))
        ]
        for i in range(len(m1))
    ]

# User Input
while True:
    print("""
Matrix Calculator:
1. Add Vectors
2. Scalar Multiplication
3. Multiply Matrix
4. Exit
""")
    user_input = int(input("Enter your choice: "))
    if user_input == 4:
        print('Program Terminated')
        break
        
    elif user_input == 1:
        print("Vector Addition")
        vector1 = list(map(int, input("Enter the first vector (space-separated): ").split()))
        vector2 = list(map(int, input("Enter the second vector (space-separated): ").split()))
        try:
            ans = add_vectors(vector1, vector2)
            print("Result:", ans)
        except ValueError as e:
            print(e)
            
    elif user_input == 2:
        print("Scalar Multiplication")
        scalar = int(input("Enter the scalar value: "))
        vector = list(map(int, input("Enter the vector (space-separated): ").split()))
        ans = scalar_matrix(scalar, vector)
        print("Result:", ans)
        
    elif user_input == 3:
        print("Matrix Multiplication")
        rows1 = int(input("Enter the number of rows for the first matrix: "))
        print("Enter the first matrix row by row (space-separated):")
        matrix1 = [list(map(int, input().split())) for _ in range(rows1)]

        rows2 = int(input("Enter the number of rows for the second matrix: "))
        print("Enter the second matrix row by row (space-separated):")
        matrix2 = [list(map(int, input().split())) for _ in range(rows2)]

        try:
            result = multiply_matrix(matrix1, matrix2)
            print("Result:")
            for row in result:
                print(row)
        except ValueError as e:
            print(e)
    else:
        print("Invalid choice. Please try again.")
```
