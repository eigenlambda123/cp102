## Problem2

Create a program that will accept Prelim, Midterm, Semi-Final, Final, Quiz, and Project Grade. 
Compute the final grade: FG = (prelime x 15%) + (midterm x 15%) + (semifinal x 15%) + (final x 15%) + (quiz x 25%) + (project x 15%)
Display a remark Congratulations! You passed the course if final grade is 75 and above; otherwise display Sorry, you failed

```python
# Store Grades
grades = {
    "prelim": None,
    "midterm": None,
    "semis": None,
    "finals": None,
    "quiz": None,
    "project": None,
}

# Set Percentages for each grade
def calculate(grades):
    percentage = {
        "prelim": 0.15,
        "midterm": 0.15,
        "semis": 0.15,
        "finals": 0.15,
        "quiz": 0.25,
        "project": 0.15,
    }

    # Calculate grades with their respective percentages
    total_grade = sum(grades[key] * percentage[key] for key in grades)
    return total_grade


# Main Program Loop
while True:
    print("""
    Which term do you want to add your grade?
    1. Prelim
    2. Midterm
    3. Semifinal
    4. Final
    5. Quiz
    6. Project
    7. Look at Grades
    (Write 'calculate' to calculate or 'quit' to exit)
    """)

    user_choices = {
        '1': 'prelim',
        '2': 'midterm',
        '3': 'semis',
        '4': 'finals',
        '5': 'quiz',
        '6': 'project',
    }

    user_input = input(": ").strip().lower()

    if user_input == 'calculate':
        # Ensure all grades are filled
        if None in grades.values():
            print("Please fill in all grades before calculating.")
        else:
            total_grade = calculate(grades)
            print(f"\nFinal Grade: {total_grade:.2f}%")
            if total_grade < 75:
                print("Sorry, you failed.")
            else:
                print("Congratulations! You passed the course!")
            break

    elif user_input == '7':
        print("\nGrades Summary:")
        for key, value in grades.items():
            status = "Not Entered" if value is None else value
            print(f"{key.capitalize()}: {status}")
        print()

    elif user_input == 'quit':
        print("Program terminated.")
        break

    elif user_input in user_choices:
        term = user_choices[user_input]
        try:
            grade = float(input(f"Enter grade for {term.capitalize()}: "))
            if 0 <= grade <= 100:
                grades[term] = grade
                print(f"{term.capitalize()} updated to {grade}")
            else:
                print("Please enter a grade between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

    else:
        print("Invalid choice. Please select a valid option.")
```

## Problem5

Input a number and display if the number is a Prime Number or Not a Prime Number

```python
num = int(input("Give Num: "))

if num <= 1:
    print(f"{num} is not a prime number")
elif num == 2:
    print(f"{num} is a prime number")  # Exception
else:
    is_prime = True  # Assume the number is prime unless proven otherwise
    for i in range(2, int(num**0.5) + 1):  # (Formula) Check divisors up to the square root of the number
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number")
```

## Problem6

Display a series of prime numbers based on user input.

```python
user_num = int(input("Give num: "))

# Prime Number Calculator
def prime_number_checker(num):
    if num <= 1:
        return False
    for i in range(2, (num//2)+1):
        if (num%i) == 0:
            return False
    return True
    
# Generator
def prime_number_generator(user_num):
    prime_numbers = []
    for i in range(1, user_num+1):
        if prime_number_checker(i):
            prime_numbers.append(i)
    return prime_numbers
    
prime_number_generator(user_num)
```

## Problem8

Accept integer number and store it to a list until user input is 0.
Display the content of the list and size of the list.
Display the highest and lowest numbers.

```python
nums = []

# Input numbers from the user
while True:
    try:
        user_num = int(input("Give Num (0 to stop): "))
        if user_num == 0:
            break
        nums.append(user_num)
    except ValueError:
        print("Please enter a valid integer.")

# Get Highest Number
def calculate_max(nums):
    maxim = nums[0]  # Set max to first index
    for i in nums:
        if i > maxim:
            maxim = i
    return maxim

# Get Lowest Number
def calculate_min(nums):
    lowest = nums[0]  # Set min to first index
    for i in nums:
        if i < lowest:
            lowest = i
    return lowest

print(f'Numbers: {nums}')
print(f'Size of list: {len(nums)}')
print(f'Highest number: {calculate_max(nums)}')
print(f'Lowest number: {calculate_min(nums)}')
```

**Matrix Calculator (adding vectors, scalars, matrix multiplication)**

```python

def add_vectors(v1, v2):
    if len(v1) != len(v2):
        raise ValueError("Vectors must be the same length.")
    return [v1[i] + v2[i] for i in range(len(v1))]

def scalar_matrix(scalar, vector):
    return [scalar * x for x in vector]

def multiply_matrix(m1, m2):
    if len(m1[0]) != len(m2):
        raise ValueError("Number of columns in the first matrix must equal the number of rows in the second matrix.")
    return [
        [
            sum(m1[i][k] * m2[k][j] for k in range(len(m2)))
            for j in range(len(m2[0]))
        ]
        for i in range(len(m1))
    ]

# User Input
while True:
    print("""
Matrix Calculator:
1. Add Vectors
2. Scalar Multiplication
3. Multiply Matrix
4. Exit
""")
    user_input = int(input("Enter your choice: "))
    if user_input == 4:
        print('Program Terminated')
        break
        
    elif user_input == 1:
        print("Vector Addition")
        vector1 = list(map(int, input("Enter the first vector (add space per number to seperate): ").split()))
        vector2 = list(map(int, input("Enter the second vector (add space per number to seperate): ").split()))
        try:
            ans = add_vectors(vector1, vector2)
            print("Result:", ans)
        except ValueError as e:
            print(e)
            
    elif user_input == 2:
        print("Scalar Multiplication")
        scalar = int(input("Enter the scalar value: "))
        vector = list(map(int, input("Enter the vector (add space per number to seperate): ").split()))
        ans = scalar_matrix(scalar, vector)
        print("Result:", ans)
        
    elif user_input == 3:
        print("Matrix Multiplication")
        rows1 = int(input("Enter the number of rows for the first matrix: "))
        print("Enter the first matrix row by row (add space per number to seperate): ")
        matrix1 = [list(map(int, input().split())) for _ in range(rows1)]

        rows2 = int(input("Enter the number of rows for the second matrix: "))
        print("Enter the second matrix row by row (add space per number to seperate): ")
        matrix2 = [list(map(int, input().split())) for _ in range(rows2)]

        try:
            result = multiply_matrix(matrix1, matrix2)
            print("Result:")
            for row in result:
                print(row)
        except ValueError as e:
            print(e)
    else:
        print("Error")
```

## Problem14

Accept string input.
Display if the string is Palindrome or Not a Palindrome

```python
user_input = input("Write anything: ").lower()

def palindrome(user_input):
    # Remove non alphanumeric characters and spaces
    cleaned_input = ''.join(char for char in user_input if char.isalnum())
    
    pointer1 = 0
    pointer2 = len(cleaned_input) - 1

    while pointer1 < pointer2:
        if cleaned_input[pointer1] != cleaned_input[pointer2]:
            return 'The word you gave is not a palindrome'
        pointer1 += 1
        pointer2 -= 1

    return 'The word you gave is a palindrome'

print(palindrome(user_input))
```

```python
import random
# Create Functions needed for game conditons

# void func for coding practice
def practice_coding(user):
    print("You practice coding!")
    user['coding'] = min(100, user['coding'] + 10)  # add coding skill by 10
    user['energy'] = max(0, user['energy'] - 20)  # Decrease energy by 20
    user['stress'] = min(100, user['stress'] + 5)  # Increase stress by 5

# for taking a break
def take_break(user):
    print("You take a well-deserved break.")
    user['happiness'] = min(100, user['happiness'] + 15)  # add happiness by 15
    user['energy'] = min(100, user['energy'] + 10)  # add energy by 10
    user['stress'] = max(0, user['stress'] - 5)  # Reduce stress by 5

# for attending a workshop
def attend_workshop(user):
    print("You attend a coding workshop.")
    user['coding'] = min(100, user['coding'] + 15)  # add coding skill by 15
    user['motivation'] = min(100, user['motivation'] + 10)  # add motivation by 10
    user['stress'] = min(100, user['stress'] + 10)  # add stress by 10
    user['energy'] = max(0, user['energy'] - 30)  # add energy by 30

#  for going for a walk
def go_for_walk(user):
    print("You go for a walk.")
    user['happiness'] = min(100, user['happiness'] + 10)  # add happiness by 10
    user['stress'] = max(0, user['stress'] - 10)  # decrease stress by 10

#  for balancing time
def balance_time(user):
    print("You focus on balancing work and life.")
    user['happiness'] = min(100, user['happiness'] + 5)  # add happiness by 5
    user['motivation'] = min(100, user['motivation'] + 5)  # add motivation by 5
    user['stress'] = max(0, user['stress'] - 5)  # decrease stress by 5

# function with return value to check the user stats
def check_status(user):
    stats = f"""
Your Status:
Coding Skill: {user['coding']}%
Happiness: {user['happiness']}%
Energy: {user['energy']}%
Motivation: {user['motivation']}%
Stress Level: {user['stress']}%
    """
    return stats

# -----------------------------------------------------------------------------------------
# -----------------------------------------------------------------------------------------

# Create Game
print("Welcome to the Coding Career Simulator!")
username = input("Enter your name: ")

# user base stats
user = {
    'coding': 50,         
    'happiness': 50,     
    'energy': 50,         
    'motivation': 50,     
    'stress': 20         
}

# game loop
while True:
    print("What would you like to do?")
    print("1. Practice Coding")
    print("2. Take a Break")
    print("3. Attend a Workshop")
    print("4. Go for a Walk")
    print("5. Balance Your Time")
    print("6. Check your status")
    print("7. Quit")
    
    user_choice = input("Enter 1, 2, 3, 4, 5, 6, or 7: ")
    
    if user_choice == "1":
        practice_coding(user)  
    elif user_choice == "2":
        take_break(user)  
    elif user_choice == "3":
        attend_workshop(user)  
    elif user_choice == "4":
        go_for_walk(user)  
    elif user_choice == "5":
        balance_time(user)  
    elif user_choice == "6":
        print(check_status(user)) 
    elif user_choice == "7":
        print("Thanks for playing the game")
        break  
    else:
        print("Error, Please Try Again")
    
    # randomly decrease user stats over time 
    user['energy'] = max(0, user['energy'] - random.randint(1, 5))  
    user['happiness'] = max(0, user['happiness'] - random.randint(1, 3))  
    user['motivation'] = max(0, user['motivation'] - random.randint(1, 3))  

    # game over conditions
    if user['energy'] <= 0:
        print("You're too tired to continue! Game over!")
        break
    elif user['happiness'] <= 0:
        print("You're too stressed out! Game over!")
        break
    elif user['motivation'] <= 0:
        print("You've lost all motivation! Game over!")
        break
    elif user['coding'] >= 100:
        print("Congratulations! You've mastered coding! You win!")
        break
```

## Problem16

Create a Python program that will demonstrate lambda.

 **Matrix Calculator (add, multiply, Transpose)**

```python
import numpy as np

# Lambda for matrix addition
matrix_add = lambda A, B: np.add(A, B)

# Lambda for matrix multiplication
matrix_multiply = lambda A, B: np.dot(A, B)

# Lambda for matrix transpose
matrix_transpose = lambda A: np.transpose(A)


def get_matrix_input():
    rows = int(input("Enter the number of rows: "))
    matrix = []
    for i in range(rows):
        row = list(map(int, input(f"Enter the elements of row {i + 1} (separate by spaces): ").split()))
        matrix.append(row)
    return np.array(matrix)

while True:
    print("""
Choose an operation:
1. Matrix Addition
2. Matrix Multiplication
3. Matrix Transpose
4. Exit
    """)

    user_input = input("Enter your choice (1/2/3/4): ")

    # Matrix Addition
    if user_input == '1':
        print("Matrix Addition selected")
        print("Enter Matrix A:")
        A = get_matrix_input()
        print("Enter Matrix B:")
        B = get_matrix_input()
        
        if A.shape == B.shape:
            result = matrix_add(A, B)
            print(f"Matrix A: \n{A}")
            print(f"Matrix B: \n{B}")
            print(f"Result of Addition: \n{result}")
        else:
            print("Error: Matrices must have the same dimensions for addition.")

    # Matrix Multiplication
    elif user_input == '2':
        print("Matrix Multiplication selected")
        print("Enter Matrix A:")
        A = get_matrix_input()
        print("Enter Matrix B:")
        B = get_matrix_input()
        
        if A.shape[1] == B.shape[0]:
            result = matrix_multiply(A, B)
            print(f"Matrix A: \n{A}")
            print(f"Matrix B: \n{B}")
            print(f"Result of Multiplication: \n{result}")
        else:
            print("Error! Number of columns in Matrix A must equal the number of rows in Matrix B")

    # Matrix Transpose
    elif user_input == '3':
        print("Matrix Transpose selected")
        print("Enter Matrix A:")
        A = get_matrix_input()
        result = matrix_transpose(A)
        print(f"Matrix A: \n{A}")
        print(f"Transpose of Matrix A: \n{result}")
    
    elif user_input == '4':
        print("Program Terminated")
        break
    
    else:
        print("Error")
```
